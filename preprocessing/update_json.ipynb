```python
f = open('drive/MyDrive/TRTPose/Dataset_Training/person_keypoints_train2017.json')
data = json.load(f)
```

```python
extra=[]
for ann in data['annotations']:
  try:
    if(len(ann['keypoints'])==24):
      for i in range(0,51):
        ann['keypoints'].append(0)
      ann['category_id']=4
      f_json['annotations'].append(ann)
    if(len(ann['keypoints'])==51):
      for i in range(0,24):
        ann['keypoints'].insert(0,0)
      ann['category_id']=4
      f_json['annotations'].append(ann)
  except:
    extra.append(ann['image_id'])
  
```

```python
print(len(set(extra)))
```

```python
a=0
for img in data['images']:
  if(img['id'] in set(extra)):
    path = os.listdir('drive/MyDrive/TRTPose/Dataset_Training/Val_Data')
    for i in path:
      if(i==img['file_name']):
        try:
          os.remove('drive/MyDrive/TRTPose/Dataset_Training/Val_Data/'+i)
        except:
          a=a+1
    continue
  img['category_ids']=4
  f_json['images'].append(img)
```
